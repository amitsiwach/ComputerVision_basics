### 3D Array

```python
# Importing libraries
import tensorflow as tf
import numpy as np
from tensorflow.keras.preprocessing import image
from PIL import Image
```

```python
# Loading the image

# test_image = image.load_img('/content/person1000_bacteria_2931.jpeg')
test_image = image.load_img('/content/person1000_bacteria_2931.jpeg', target_size = (255, 255)) # can specify the target size of array here
```

```python
# Converting image to array

test_image_array = image.img_to_array(test_image)
```

```python
# Checking the shape of immage_array

test_image_array.shape
```

```python
test_image_array
```

```python
#Converting back from image_array to Image
pil_img = tf.keras.utils.array_to_img(test_image_array)
pil_img
```

### Reading Image, converting to 1st Bytes then to numpy array

```python
# Convert image to bytes
import io
import cv2

im = cv2.imread('/content/person1000_bacteria_2931.jpeg')
im_resize = cv2.resize(im, (255, 255))

is_success, im_buf_arr = cv2.imencode(".jpg", im_resize)
byte_im = im_buf_arr.tobytes()
```

```python
# this is the byte format

byte_im
```

```python
# converting from bytes to Numpy array

image = np.array(Image.open(io.BytesIO(byte_im)))
```

```python
#

image.shape
```

```python
image
```

```python
pil_img = tf.keras.utils.array_to_img(image)
pil_img
```
